```python
import pandas as pd
import numpy as np
from polymarket_events import Polymarket_events
from datetime import timedelta
```

```python
path_to_args = 'private_kargs.json'
DEP = 200
leagues = ['NFL', 'NBA', 'NCAA', 'NHL']
```

```python
signal_list = []

for league in leagues:
    poly_league = Polymarket_events(path_to_args=path_to_args, league=league)
    signal = poly_league.run_signal()
    signal['best_bid_kelly'] = signal['best_bid_kelly'] * DEP
    signal['best_ask_kelly'] = signal['best_ask_kelly'] * DEP
    signal_list.append(signal)

signal_all = pd.concat(signal_list)
```

```python
signal_all[signal_all['best_bid'] < signal_all['min']].sort_values(by='best_bid_kelly', ascending=False)
```

```python
betting_table = signal_all[signal_all['best_bid_kelly'] > 0]
betting_table = betting_table[betting_table['best_bid'] < betting_table['min']]
betting_table.sort_index(level='commence_time')
```
